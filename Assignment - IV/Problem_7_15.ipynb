To generate a signal u of length m = 50, with each entry a random value that is either -1 or +1

```julia
u = rand([-1,1], 50)
print(u)
```

Plot u

```julia
using Plots
Plots.pyplot()
plot(u, framestyle = :zerolines, st = :sticks, label = "u", xlabel = "t", ylabel = "u (t)")
```

Create c = (1, 0.7, -0.3)

```julia
c = [1, 0.7, -0.3]
```

y = c * u

```julia
using DSP
y = conv(c, u)
print(y)
```

Plot y

```julia
plot(y, framestyle = :zerolines, st = :sticks, label = "y", xlabel = "t", ylabel = "y (t)")
```

h = (0.9, -0.5, 0.5, -0.4, 0.3, -0.3, 0.2, -0.1)

```julia
h = [0.9, -0.5, 0.5, -0.4, 0.3, -0.3, 0.2, -0.1]
```

z = h * y

```julia
z = conv(h, y)
print(z)
```

Plot z

```julia
plot(z, framestyle = :zerolines, st = :sticks, label = "z", xlabel = "t", ylabel = "z (t)")
```

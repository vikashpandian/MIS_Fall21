Defining the size for $m\ and\ n$

```julia
m = 10000
n = 100
```

Creating a random matrix A with $m \times n$ elements 

```julia
A = rand(m, n)
```

Creating a random vector b with $m$ elements 

```julia
b = rand(m)
```

Finding the time of $A \backslash b$ operation (Solving least squares)

```julia
total_time = 0
@show A\b
for i in 1:100
    total_time = total_time + @elapsed A\b
end
```

```julia
@show total_time 
@show average_time = total_time / 100
```
